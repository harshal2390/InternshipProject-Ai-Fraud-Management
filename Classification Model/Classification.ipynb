```python
from ultralytics import YOLO
import os
```

```python
dataset_path = r"E:\internship project\Classification Model\Classification dataset"

for root, dirs, files in os.walk(dataset_path):
    print(root, "->", len(files), "files")
```

```python
# Load YOLOv8 classification model (you can also use yolo11n-cls.pt if downloaded)
model = YOLO("yolov8n-cls.pt")

# Train the classifier
results = model.train(
    data=dataset_path,
    epochs=15,
    imgsz=224,     # 224x224 typical for classification
    batch=8
)

print("✅ Training complete")
```

```python
trained_classifier = r"E:\internship project\runs\classify\train\weights\best.pt"
model = YOLO(trained_classifier)

print("✅ Loaded model:", trained_classifier)
```

```python
metrics = model.val()
print("✅ Validation done")
print("Top-1 Accuracy:", metrics.top1)
print("Top-5 Accuracy:", metrics.top5)
```

```python
test_image = r"E:\internship project\Classification Model\Classification dataset\val\non_aadhar\1.jpg"  # example

results = model(test_image)
results[0].show()
print("✅ Prediction complete")
```
